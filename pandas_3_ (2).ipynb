```python
import pandas as pd
ipl_data={'Team':['Rider','Devils','Devils','Kings','Kings','Kings','Kings','Kings','Riders','Royals','Royals','Riders'],'Rank':[1,2,2,3,3,4,1,1,2,4,1,2],
          'year':[2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],'points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df=pd.DataFrame(ipl_data)
print(df)
```

```python
df1= df.groupby('Rank').groups
print(df1)
```

```python
df2=df.groupby('Rank').max()
print(df2)
```

```python
df2=df.groupby('Rank').max()['points']
print(df2)
```

```python
df2=df.groupby('Team').max()['Rank']
print(df2)
```

```python
df2=df.groupby('Team').min()['Rank']
print(df2)
```

```python
import pandas as pd # Import the pandas library and alias it as 'pd'

df=pd.DataFrame(ipl_data)
grouped=df.groupby('Team')
print(grouped.count())
```

```python
df=pd.DataFrame(ipl_data)
grouped=df.groupby('Team')
score=lambda x:x+20
#score=lambda x:(x-x.mean())/x.std()*10
print(grouped.transform(score))
```

```python
grouped=df.groupby('Team')['Rank']
score=lambda x:x+20
#score=lambda x:(x-x.mean())/x.std()*10
print(grouped.transform(score))
```

```python
df=pd.DataFrame(ipl_data)
print(df.groupby('Team').filter(lambda x:len(x)>=3))
```
